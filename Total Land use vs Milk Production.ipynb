```python
# Visualization
import numpy as np
from matplotlib import style
from statistics import mean

Land_list = pf['Global Landuse per year (km2)'].values.tolist()
Milk_list = pf['Global Milk Production (pw)'].values.tolist()

xs = np.array(Land_list, dtype=np.float64)
ys = np.array(Milk_list, dtype=np.float64)

def best_fit_line_slope(xs,ys):
    slope = (((mean(xs) * mean(ys)) - mean(xs * ys)) / ((mean(xs) * mean(xs)) - mean(xs * xs)))
    return slope

def best_fit_line_y(xs,ys):
    slope = (((mean(xs) * mean(ys)) - mean(xs * ys)) / ((mean(xs) * mean(xs)) - mean(xs * xs)))
    y_value = mean(ys) - slope * mean(xs)
    return y_value

Slope = best_fit_line_slope (xs,ys)
Y_value = best_fit_line_y(xs,ys)

regression_line = [(Slope * x) + Y_value for x in xs]


pf.plot(kind='scatter', x= 'Global Landuse per year (km2)' , y= 'Global Milk Production (pw)', 
        alpha = 0.6, color = 'green', s = 40)

style.use ("seaborn")
plt.plot(xs,regression_line, label='Best Fit Line', color = 'orange', linewidth = 4)



```
