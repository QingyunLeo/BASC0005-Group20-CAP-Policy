```python
import pandas as pd
```

```python
df = pd.read_csv ('RegressionFolder/Landuse.csv')
```

```python
## Cleaning the dataset

df.drop([0,57,58,59], inplace = True)
df
```

```python
df.drop(labels = ["Unnamed: 3"], axis = 1, inplace = True)

df
```

```python
df.rename(columns={"Time": "OECD Economies", "Unnamed: 1": "Non-OECD Economies", "Unnamed: 2": "Unit"}, inplace = True)
```

```python
df['OECD Economies'].replace('Non-OECD Economies', 'NaN', inplace = True)
```

```python
 ##df.drop([40], inplace = True)

## correcting the index error after dropping 40

df.reset_index(drop=True, inplace=True)
```

```python
## cleaning the commas
df.replace(',', '', regex=True, inplace = True)
```

```python
df['1990'].dtypes
```

```python
type(df['1990'][1])
```

```python
## Sum up the global landuse for agricultural purpose in each year
import numpy as np
cols = df.columns[3:]
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df[cols] = df[cols].replace(np.nan, 0, regex=True)
df_sum = df[cols].sum()
df_sum
```

```python
df
```

```python
## Change Unit From Hectares to Meter (1 Hectares = 0.01 Kilometer square, so divided by 100)
landuse_se = df_sum.div(100).round(2)
landuse_se
```

```python
## Now we import the global milk production data 
mf = pd.read_csv ('RegressionFolder/Global milk production.csv')
```

```python
# Clean the dataset
mf.drop([0,1,3], inplace = True)
mf
```

```python
# We only need to extract the 4th line since it is the global milk production from 1990 to 2021
# To compare with Global Landuse 1990 to 2021
row = mf.iloc[1]
row.replace(',', '', regex=True, inplace = True)
row
```

```python
## Construct New Dataframe
milk_pro = pd.to_numeric(row[2:31])
milk_pro

milks_pro = pd.Series(milk_pro).values
milks_pro

## We only need to column up to 2019 to compare
```

```python
## The last year (2019) seems like an outlier, so we delete it.
landuse_final = pd.to_numeric(landuse_se[0:29])
                              
## Process with the landuse data
landuses_final = pd.Series(landuse_final).values
landuses_final
```

```python
## Create New Dataframe
qf = pd.DataFrame({'Landuse data': landuses_final, 'Global Milk Production data': milks_pro})
qf
## qf is the data of Global Landuse and Milk Production from 1990 to 2018
## 2019 was omited due to the existence of outlier with inadequate amount of data
```

```python
## Simple Linear Regression Model 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

qf.plot(kind = 'scatter', x = 'Global Milk Production data', y = 'Landuse data')
plt.show()
```

```python
## Test train split for supervised training
X_train, X_test, y_train, y_test = train_test_split(milks_pro, landuses_final)

## Test train split visualization
plt.scatter(X_train, y_train, label = 'Training Data', color = 'r', alpha = .7)
plt.scatter(X_test, y_test, label = 'Testing Data', color = 'g', alpha = .7)
plt.title ('Test Train Split')
plt.show()
```

```python
## Create Linear Regression Model and test it
LR = LinearRegression()
LR.fit(X_train.reshape(-1,1), y_train)
```

```python
prediction = LR.predict(X_test.reshape(-1,1))
plt.plot(X_test, prediction, label = 'Linear Regression', color = 'b')
plt.scatter(X_test, y_test, label = 'Actual Test Data', color = 'g', alpha = .7)
```

```python
## Score the model
LR.score(X_test.reshape(-1,1), y_test)
```
